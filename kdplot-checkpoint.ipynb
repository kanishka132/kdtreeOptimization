```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from files
kd_data = np.loadtxt('kdtree.t')  
kdomp_data = np.loadtxt('kdtree_OMP.t', dtype=float, delimiter=' ')


# Create plot
fig, ax = plt.subplots()
ax.axhline(kd_data, linestyle='--', color='b',label='kdtree')
ax.plot(kdomp_data[:,0], kdomp_data[:,1], 'g', label='kdtree_OMP') 


ax.set_xlabel('Number of threads')
ax.set_ylabel('Time (seconds)')
ax.set_title('Timing comparision between Serial and Parallel Kdtree')
ax.legend()

fig.savefig('timing_comparisions.png')
```
